<h1>RELU activation</h1>

```python
import numpy as np
inputs = [0,2,-1,3.3,-2.7,1.1,2.2,-100]
output = np.maximum(0, inputs)
print(output)
```

```python
# ReLU activation class
class ReLUActivation:
    def forward(self, inputs):
        self.output = np.maximum(0, inputs)
```

ReLU can be used in forward layers, but for output we want something that gives probabilities.
This is best done by Softmax.

```python
# Let's say we get the inputs as follows from neural network for 3 batches and 3 features
inputs = [[1, 2.5, 1.0], [2.0, 3.0, 2.0], [-1.5, 2.7, 3.0]]

# unnormalized probabilities
exp_values = np.exp(inputs - np.max(inputs, axis=1, keepdims=True))

# normalized probabilities
probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)

print(probabilities)

np.sum(probabilities, axis=1)
```

Softmax activation class

```python
class SoftmaxActivation:
    def forward(self, inputs):
        exp_values = np.exp(inputs - np.max(inputs, axis=1, keepdims=True))
        probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        self.output = probabilities
```

<h3>Let's check with spiral data</h3>

```python
from nnfs.datasets import spiral_data
```

```python
class DenseLayer:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
```

```python
X, y = spiral_data(samples=100, classes=3)

dense1 = DenseLayer(2, 3)
activation1 = ReLUActivation()

dense2 = DenseLayer(3, 3)
activation2 = SoftmaxActivation()

dense1.forward(X)
activation1.forward(dense1.output)

dense2.forward(activation1.output)
activation2.forward(dense2.output)

print(activation2.output[:5])
```
