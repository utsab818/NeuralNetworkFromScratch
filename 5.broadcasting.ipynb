Importance of keepdim in numpy.

```python
import numpy as np
inputs = [[1,2,3], [4,5,6], [7,8,9]]

print(np.sum(inputs))
```

```python
# row
print(np.sum(inputs, axis=0))
print(np.sum(inputs, axis=0).shape)

# column
print(np.sum(inputs, axis=1))
print(np.sum(inputs, axis=1).shape)
```

```python
# keepdims preserves the dimensions of the original array
print(np.sum(inputs, axis=0, keepdims=True))
print(np.sum(inputs, axis=0, keepdims=True).shape)
print("\n")
print(np.sum(inputs, axis=1, keepdims=True))
print(np.sum(inputs, axis=1, keepdims=True).shape)
```

Print maximum value according to row or column.

```python
print(np.max(inputs, axis=0))
print(np.max(inputs, axis=0).shape)
print(np.max(inputs, axis=1))
print(np.max(inputs, axis=1).shape)
```

```python
print(np.max(inputs, axis=0, keepdims=True))
print(np.max(inputs, axis=0, keepdims=True).shape)
print("\n")
print(np.max(inputs, axis=1, keepdims=True))
print(np.max(inputs, axis=1, keepdims=True).shape)
```
